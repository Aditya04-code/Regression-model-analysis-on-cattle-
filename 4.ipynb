```python
fig4, axes4 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df4['Count of protein'],kde=True,ax=axes4[0])
sns.boxplot(df4['Count of protein'],color='b',ax=axes4[1])
axes4[0].set_title('Ditribution of Count of proteins')
axes4[1].set_title('Count of proteins range')
plt.show()
```

```python
fig4, axes4 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df4['GO frequency within the dataset (%)'],kde=True,ax=axes4[0])
sns.boxplot(df4['GO frequency within the dataset (%)'],color='b',ax=axes4[1])
axes4[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes4[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig4, axes4 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df4['GO frequency within the genome (%)'],kde=True,ax=axes4[0])
sns.boxplot(df4['GO frequency within the genome (%)'],color='b',ax=axes4[1])
axes4[0].set_title('Ditribution of GO frequency within the genome (%)')
axes4[1].set_title('GO frequency within the genome (%) range')
plt.show()
```

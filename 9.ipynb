```python
#setting structure for graphs
sns.set(style="whitegrid",palette='muted',color_codes=True)
```

```python
fig9, axes9 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df9['Count of protein'],kde=True,ax=axes9[0])
sns.boxplot(df9['Count of protein'],color='b',ax=axes9[1])
axes9[0].set_title('Ditribution of Count of proteins')
axes9[1].set_title('Count of proteins range')
plt.show()
```

```python
fig9, axes9 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df9['GO frequency within the dataset (%)'],kde=True,ax=axes9[0])
sns.boxplot(df9['GO frequency within the dataset (%)'],color='b',ax=axes9[1])
axes9[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes9[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig9, axes9 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df9['GO frequency within the genome (%)'],kde=True,ax=axes9[0])
sns.boxplot(df9['GO frequency within the genome (%)'],color='b',ax=axes9[1])
axes9[0].set_title('Ditribution of GO frequency within the genome (%)')
axes9[1].set_title('GO frequency within the genome (%) range')
plt.show()
```

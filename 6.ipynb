```python
fig6, axes6 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df6['Count of protein'],kde=True,ax=axes6[0])
sns.boxplot(df6['Count of protein'],color='b',ax=axes6[1])
axes6[0].set_title('Ditribution of Count of proteins')
axes6[1].set_title('Count of proteins range')
plt.show()
```

```python
fig6, axes6 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df6['GO frequency within the dataset (%)'],kde=True,ax=axes6[0])
sns.boxplot(df6['GO frequency within the dataset (%)'],color='b',ax=axes6[1])
axes6[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes6[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig6, axes6 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df6['GO frequency within the genome (%)'],kde=True,ax=axes6[0])
sns.boxplot(df6['GO frequency within the genome (%)'],color='b',ax=axes6[1])
axes6[0].set_title('Ditribution of GO frequency within the genome (%)')
axes6[1].set_title('GO frequency within the genome (%) range')
plt.show()
```

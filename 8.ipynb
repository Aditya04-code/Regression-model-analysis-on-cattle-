```python
#setting structure for graphs
sns.set(style="whitegrid",palette='muted',color_codes=True)
```

```python
fig8, axes8 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df8['Count of protein'],kde=True,ax=axes8[0])
sns.boxplot(df8['Count of protein'],color='b',ax=axes8[1])
axes8[0].set_title('Ditribution of Count of proteins')
axes8[1].set_title('Count of proteins range')
plt.show()
```

```python
fig8, axes8 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df8['GO frequency within the dataset (%)'],kde=True,ax=axes8[0])
sns.boxplot(df8['GO frequency within the dataset (%)'],color='b',ax=axes8[1])
axes8[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes8[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig8, axes8 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df8['GO frequency within the genome (%)'],kde=True,ax=axes8[0])
sns.boxplot(df8['GO frequency within the genome (%)'],color='b',ax=axes8[1])
axes8[0].set_title('Ditribution of GO frequency within the genome (%)')
axes8[1].set_title('GO frequency within the genome (%) range')
plt.show()
```

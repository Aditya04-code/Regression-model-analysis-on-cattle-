```python
sns.pairplot(data=df2)  
```

```python
fig2, axes2 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df2['Count of protein'],kde=True,ax=axes2[0])
sns.boxplot(df2['Count of protein'],color='b',ax=axes2[1])
axes2[0].set_title('Ditribution of Count of proteins')
axes2[1].set_title('Count of proteins range')
plt.show()
```

```python
fig2, axes2 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df2['GO frequency within the dataset (%)'],kde=True,ax=axes2[0])
sns.boxplot(df2['GO frequency within the dataset (%)'],color='b',ax=axes2[1])
axes2[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes2[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig2, axes2 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df2['GO frequency within the genome (%)'],kde=True,ax=axes2[0])
sns.boxplot(df2['GO frequency within the genome (%)'],color='b',ax=axes2[1])
axes2[0].set_title('Ditribution of GO frequency within the genome (%)')
axes2[1].set_title('GO frequency within the genome (%) range')
plt.show()
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr(),annot=True)
```

```python
print("GO frequency within the dataset (%) : skewness %f",df2['GO frequency within the dataset (%)'].skew())
print("GO frequency within the dataset (%) : kurtosis %f",df2['GO frequency within the dataset (%)'].kurt())
print("Count of protein : skewness %f",df2['Count of protein'].skew())
print("Count of protein : kurtosis %f",df2['Count of protein'].kurt())
print("GO frequency within the genome (%) : skewness %f",df2['GO frequency within the genome (%)'].skew())
print("GO frequency within the genome (%) : kurtosis %f",df2['GO frequency within the genome (%)'].kurt())
```

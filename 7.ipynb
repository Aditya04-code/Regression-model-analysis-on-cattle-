```python
fig7, axes7 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df7['Count of protein'],kde=True,ax=axes7[0])
sns.boxplot(df7['Count of protein'],color='b',ax=axes7[1])
axes7[0].set_title('Ditribution of Count of proteins')
axes7[1].set_title('Count of proteins range')
plt.show()
```

```python
fig7, axes7 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df7['GO frequency within the dataset (%)'],kde=True,ax=axes7[0])
sns.boxplot(df7['GO frequency within the dataset (%)'],color='b',ax=axes7[1])
axes7[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes7[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig7, axes7 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df7['GO frequency within the genome (%)'],kde=True,ax=axes7[0])
sns.boxplot(df7['GO frequency within the genome (%)'],color='b',ax=axes7[1])
axes7[0].set_title('Ditribution of GO frequency within the genome (%)')
axes7[1].set_title('GO frequency within the genome (%) range')
plt.show()
```

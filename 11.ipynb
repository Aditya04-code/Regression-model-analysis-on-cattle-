```python
#setting structure for graphs
sns.set(style="whitegrid",palette='muted',color_codes=True)
```

```python
fig11, axes11 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df11['Count of protein'],kde=True,ax=axes11[0])
sns.boxplot(df11['Count of protein'],color='b',ax=axes11[1])
axes11[0].set_title('Ditribution of Count of proteins')
axes11[1].set_title('Count of proteins range')
plt.show()
```

```python
fig11, axes11 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df11['GO frequency within the dataset (%)'],kde=True,ax=axes11[0])
sns.boxplot(df11['GO frequency within the dataset (%)'],color='b',ax=axes11[1])
axes11[0].set_title('Ditribution of GO frequency within the dataset (%)')
axes11[1].set_title('GO frequency within the dataset (%) range')
plt.show()
```

```python
fig11, axes11 = plt.subplots(1,2, figsize=(12,6))
sns.distplot(x=df11['GO frequency within the genome (%)'],kde=True,ax=axes11[0])
sns.boxplot(df11['GO frequency within the genome (%)'],color='b',ax=axes11[1])
axes11[0].set_title('Ditribution of GO frequency within the genome (%)')
axes11[1].set_title('GO frequency within the genome (%) range')
plt.show()
```
